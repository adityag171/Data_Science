```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize= (9,6))
plt.bar(x = df_used_cars["Brand"],
       height=df_used_cars["Cars Listings"],
       color= ["midnightblue"])
plt.xticks(rotation = 45)
plt.title("Cars Listing by Brand", fontsize=16, fontweight = "bold")
plt.ylabel("Number of Listings", fontsize = 16)
plt.show()
plt.savefig("Used Cars Bar.png")
```

```python
df_fuel_engine = pd.read_csv("pie_chart_data.csv")
```

```python
df_fuel_engine
```

```python
sns.set_palette('colorblind')
```

```python
plt.pie(x = df_fuel_engine["Number of Cars"],
       labels = df_fuel_engine["Engine Fuel Type"].values,
       autopct= "%.2f%%",
       textprops={"size" : "x-large",
                  'fontweight' : 'bold',
                  'rotation':'30',
                  'color' : 'w'})
plt.title("car by engine fuel type", fontsize = 18, fontweight = 'bold')
plt.legend()
plt.show()
```
